```python
import glob
# find the all wav files in th dir
wav_audios = glob.glob('**/*.wav',recursive = True)
mp3_audios = glob.glob('**/*.mp3',recursive = True)
audios = wav_audios+mp3_audios
len(audios) 
```

```python
# import the libraries
import matplotlib.pyplot as plt
import librosa
import numpy as np
import librosa.display
# let's take first 5 seconds of this file as example
y,sr = librosa.load(audios[0], duration=5)

import IPython.display as ipd
ipd.Audio(y, rate=sr)
```

3 letters have been spoken in this audio clip: 'a', 'b', 'c'

```python
# let's plot the waveform
plt.figure(figsize=(14, 5))
librosa.display.waveplot(y, sr=sr)
```

```python
# let's plot the spectrogram of these 3 letters
D = librosa.stft(y)  # STFT of y
S_db = librosa.amplitude_to_db(np.abs(D), ref=np.max)
 
plt.figure(figsize=(14, 5))
librosa.display.specshow(S_db)
plt.colorbar()
```

You can observe that the formants for these three letters here.

Formants are distinctive frequency components of the acoustic signal corresponding to the highlighted horizontal bars on a spectrogram (frequency band), representing spectral peaks, usually of vowels.
